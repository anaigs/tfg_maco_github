Ejemplo distribucion de Poisson

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import poisson

# Lista de valores de lambda
lambdas = [0.7, 2.5, 10]
titles = [r'$\lambda = 0.7$', r'$\lambda = 2.5$', r'$\lambda = 10$']
x_ranges = {
    0.7: np.arange(0, 8),
    2.5: np.arange(0, 12),
    10: np.arange(0, 21)
}

# Crear figura con 3 subgráficas en una fila
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

for i, lam in enumerate(lambdas):
    x = x_ranges[lam]
    y = poisson.pmf(x, mu=lam)

    axes[i].bar(x, y, color='skyblue', edgecolor='black')
    axes[i].set_title(f'Distribución de Poisson ({titles[i]})')
    axes[i].set_xlabel('k (número de eventos)')
    axes[i].set_ylabel('P(X = k)')
    axes[i].grid(True)

plt.tight_layout()
plt.show()
```
